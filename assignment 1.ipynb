3.Largest number

```python
a=int(input("Enter 1st integer "))
b=int(input("Enter 2nd integer "))
c=int(input("Enter 3rd integer "))
d=int(input("Enter 4th integer "))
if a>b and a>c and a>d:
  print("{} is the Largest number ".format(a))
elif b>a and b>c and b>d:
  print("{} is the Largest number ".format(b))
elif c>a and c>b and c>d:
  print("{} is the Largest number ".format(c))
else:
  print("{} is the Largest number ".format(d))
```

7.Reverse list

```python
my_list=[10,40,30,70]
new_list=[]
for i in range(len(my_list)):
  new_list.insert(i,my_list[-1])
  my_list.pop(-1)
print(new_list)
```

8.Pattern

```python
n=4
for i in range(0,n):
  for j in range(0,i+1):
    print("*",end=" ")
  print()
```

```python
n=5
for i in range(n):
    print("*"*(i+1))
for i in range(n,-1,-1):
    print("*"*(i+1))
```

12.Exponent

```python
def exponent(base,exp):
  if exp==1:
    return 1
  else:
    return(base**exp)
base=int(input("Enter an integer"))
exp=int(input("Enter an integer"))
exponent(base,exp)
```

13. Sum of cubes

```python
def sum_cubes(n):
  sum=0
  for i in range(1,n):
    sum+=i*i*i
  return sum
n=int(input("Enter an integer"))
sum_cubes(n)
```

15.Factorial

```python
def factorial(n):
  if n==0:
    return 1
  elif n>0:
    return (n*factorial(n-1))
  else:
    return "error"
n=int(input("Enter an integer "))
factorial(n)
```
